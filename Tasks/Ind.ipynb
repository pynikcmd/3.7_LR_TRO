## Задание

Условие. Считать цветное изображение и выполнить вывод следующих получившихся изображений:
1. Конвертировать цвета изображения инверсивно
2. Сделать цвета более яркими и насыщенными
3. Убрать красный и синий цвет

```python
получившихсяimport cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# считываем изображение
img = cv2.imread('image_ind.jpg')

# создаем копии изображения
inv_img = cv2.bitwise_not(img)
bright_img = cv2.convertScaleAbs(img, alpha=1.5, beta=0)
no_red_blue_img = np.copy(img)
no_red_blue_img[:, :, 0] = 0  # убираем красный цвет
no_red_blue_img[:, :, 2] = 0  # убираем синий цвет

# выводим все изображения вместе с оригиналом
fig, axs = plt.subplots(2, 2)
fig.suptitle('Image transformations')
axs[0, 0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axs[0, 0].set_title('Original')
axs[0, 1].imshow(cv2.cvtColor(inv_img, cv2.COLOR_BGR2RGB))
axs[0, 1].set_title('Inverted')
axs[1, 0].imshow(cv2.cvtColor(bright_img, cv2.COLOR_BGR2RGB))
axs[1, 0].set_title('Brightened')
axs[1, 1].imshow(cv2.cvtColor(no_red_blue_img, cv2.COLOR_BGR2RGB))
axs[1, 1].set_title('No red and blue')
for ax in axs.flat:
    ax.label_outer()
plt.show()
```
